```python
# Creating Strings:
single_quoted = 'Hello, world!'
double_quoted = "Hello, world!"
triple_quoted = '''Hello, world!'''
print(single_quoted)
print(double_quoted)
print(triple_quoted)

# Accessing Characters in Strings:
s = "Hello"
print(s[0])

# String Slicing:
s = "Hello, world!"
print(s[0:5]) 

# String Concatenation:
s1 = "Hello"
s2 = "world"
s3 = s1 + " " + s2
print(s3)  

# Strings Method

print(len(s1))
print(s1.lower())
print(s1.upper())
```

```python
# Numeric Types:
integer = 42
Float = 3.14
complex_num = 2 + 3j
print(integer)
print(Float)
print(complex_num)

#Sequence Types:
List = [1, 2, 3, 4, 5]
Tuple = (1, 2, 3, 4, 5)
Range = range(0, 10)
print(List)
print(Tuple)
print(list(Range))

# text type
String = "hello world"
print(String)

#mapping type
dictionary = {'name': 'Alice', 'age': 30}
print(dictionary)

#Set Type:
Set = {1, 2, 3, 4, 5}
print(Set)

#Boolean type
Booleant = True
Booleanf = False
print(Booleant)
print(Booleanf)
```

```python
#ForLoop

my_list = [1, 2, 3, 4, 5]
for item in my_list:
    print(item)

my_string = "Hello"
for char in my_string:
    print(char)

for i in range(5):
    print(i)

#While Loop

num = 0
while num < 5:
    print(num)
    num += 1

num = 0
while True:
    print(num)
    num += 1
    if num >= 5:
        break



# Nested loops
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#If Statement:
x = 5
if x > 0:
    print("x is positive")

# If-else Statement:
x = 5
if x % 2 == 0:
    print("x is even")
else:
    print("x is odd")

# If-elif-else Statement:
x = 5
if x > 0:
    print("x is positive")
elif x < 0:
    print("x is negative")
else:
    print("x is zero")

# Nested If Statement:
x = 5
if x > 0:
    if x % 2 == 0:
        print("x is positive and even")
    else:
        print("x is positive and odd")
else:
    print("x is not positive")
```

```python
# Function without Parameters:
def greet():
    print("Hello, world!")

greet()  

# Function with Parameters:
def greet(name):
    print("Hello,", name)

greet("Alice")

# Function with Default Parameters:
def greet(name="world"):
    print("Hello,", name)

greet()  
greet("Alice") 

#Function with Return Value:
def add(a, b):
    return a + b

result = add(3, 5)
print(result) 

#Recursive Function:
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

```python
# Simple Class:
class MyClass:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello,", self.name)

obj = MyClass("Alice")
obj.greet() 

# Class with Class Variable and Instance Variable:
class Dog:
    species = "Canis"

    def __init__(self, name):
        self.name = name

dog1 = Dog("Buddy")
dog2 = Dog("Molly")

print(dog1.name)  
print(dog2.name) 
print(dog1.species) 
print(dog2.species) 

# Class with Private Members:
class MyClass:
    def __init__(self):
        self.__private_var = 42

    def get_private_var(self):
        return self.__private_var

obj = MyClass()
print(obj.get_private_var())
```
